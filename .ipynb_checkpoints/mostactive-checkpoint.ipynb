```python
import requests
import pandas as pd 

url = "https://yahoo-finance-api-data.p.rapidapi.com/market/equity"

headers = {
	"x-rapidapi-key": "8dc7f3f670mshf5316a9ce715732p14098ajsnf6ee01f4a31e",
	"x-rapidapi-host": "yahoo-finance-api-data.p.rapidapi.com"
}

all_data = []
for i in range(0, 180, 20):
    querystring = {"screenerId": "MOST_ACTIVES", "region": "us", "offset": i, "limit": "20"}
    response = requests.get(url, headers=headers, params=querystring)
    if response.status_code == 200:
        status_data = response.json()
        data = status_data['data']['quotes']
        data1 = pd.DataFrame(data)
        all_data.append(data1)
    else:
        print('Error')
df = pd.concat(all_data, ignore_index=True)
```
